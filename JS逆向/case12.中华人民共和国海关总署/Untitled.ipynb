```python
import requests


headers = {
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
    "Accept-Language": "zh-CN,zh;q=0.9,en;q=0.8",
    "Cache-Control": "no-cache",
    "Connection": "keep-alive",
    "Pragma": "no-cache",
    "Referer": "http://www.customs.gov.cn/customs/302249/302266/08654b53-1.html",
    "Upgrade-Insecure-Requests": "1",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36"
}
url = "http://www.customs.gov.cn/customs/302249/302266/08654b53-1.html"
response = requests.get(url, headers=headers, verify=False)

print(response.text)
print(response)
```

```python
dict_cookie = {[*response.cookies][0].name: [*response.cookies][0].value}

dict_cookie
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, 'html5lib')

print(soup.prettify())
```

```python
from parsel import Selector
```

```python
selector = Selector(text=str(soup))

selector
```

```python
U07cUYgw9lbI_content = selector.xpath('//meta[@id="U07cUYgw9lbI"]/@content').get()

U07cUYgw9lbI_content
```

```python
js_1 = selector.xpath('//script[@type="text/javascript"]')[0].xpath('./text()').get()

js_1
```

```python
url_js = f'''http://www.customs.gov.cn{selector.xpath('//script[@type="text/javascript"]')[1].xpath('./@src').get()}'''

url_js
```

```python
response = requests.get(url_js, headers=headers, verify=False)

print(response.text)
print(response)
```

```python
with open('./js.js', 'r', encoding='utf-8') as f:
    js = f.read()
```

```python
import subprocess
```

```python
js = js.replace('Lennon_U07cUYgw9lbI_content', U07cUYgw9lbI_content).replace('Lennon_js_1', js_1).replace('Lennon_js_2', response.text)
with open('./now.js', 'w', encoding='utf-8') as f:
    f.write(js)
result = subprocess.run(['node', './now.js'], capture_output=True, text=True, encoding='utf-8')

result.stdout
```

```python
key, value = result.stdout.split(';')[0].split('=', 1)
dict_cookie[key] = value

dict_cookie
```

```python
response = requests.get(url, headers=headers, cookies=dict_cookie, verify=False)

print(response.text)
print(response)
```
